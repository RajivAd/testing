```python
plt.figure(figsize(10,8))
fb1['Close'].plot()
plt.show()
```

```python
fb1['price1'] = fb1['Close'].shift(-1)
fb1
```

```python
fb1['price_diff'] = fb1['price1'] - fb1['Close']
```

```python
fb1
```

```python
fb1['return_daily'] = fb1['price_diff']/fb1['Close']
fb1
```

```python
fb1['derection'] = [1 if fb1['price_diff'].loc[ei]> 0 else 0 for ei in fb1.index]
fb1
```

```python
fb1['derection_open'] = [1 if fb1['Open'].loc[ei]>0 else 0 for ei in fb1.index]
fb1
```

```python
fb1['direction_high'] = ['+ve' if fb1['High'].loc[ei]>0 else '+ve' for ei in fb1.index]
fb1
```

```python
# moving average calculation
```

```python
fb1['Moving_avg'] = (fb1['Close']+fb1['Close'].shift(1)+fb1['Close'].shift(2))/3
fb1
```

```python
fb1['MA40'] = fb1['Close'].rolling(40).mean()
fb1
```

```python
plt.figure(figsize=(10,8))
fb1['Close'].plot()
fb1['Moving_avg'].plot()
fb1['Close'].rolling(40).mean().plot(label = 'MA-40')
plt.legend()
plt.show()
```

```python
fb1['MA_10'] = fb1['Close'].rolling(10).mean()
fb1
```
